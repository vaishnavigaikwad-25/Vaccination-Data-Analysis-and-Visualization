## Create Normalized SQL Tables


Generate and execute SQL DDL statements to create tables based on the defined schema, including separate tables for entities like vaccines, diseases, countries, and years. Ensure primary and foreign keys are correctly implemented to enforce referential integrity.


```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_who_region (
    who_region_code TEXT NOT NULL PRIMARY KEY,
    who_region_name TEXT NOT NULL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_group_type (
    group_type_code TEXT NOT NULL PRIMARY KEY
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_country (
    country_code TEXT NOT NULL PRIMARY KEY,
    country_name TEXT NOT NULL,
    who_region_code TEXT NOT NULL,
    group_type_code TEXT NOT NULL,
    FOREIGN KEY (who_region_code) REFERENCES dim_who_region(who_region_code),
    FOREIGN KEY (group_type_code) REFERENCES dim_group_type(group_type_code)
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_year (
    year_id INTEGER NOT NULL PRIMARY KEY
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_antigen (
    antigen_code TEXT NOT NULL PRIMARY KEY,
    antigen_description TEXT NOT NULL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_coverage_category (
    coverage_category_code TEXT NOT NULL PRIMARY KEY,
    coverage_category_description TEXT NOT NULL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_disease (
    disease_code TEXT NOT NULL PRIMARY KEY,
    disease_description TEXT NOT NULL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_denominator (
    denominator_description TEXT NOT NULL PRIMARY KEY
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_vaccine_intro_product (
    vaccine_intro_product_name TEXT NOT NULL PRIMARY KEY
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_intro_status (
    intro_status_code TEXT NOT NULL PRIMARY KEY
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_vaccine_schedule (
    vaccine_schedule_code TEXT NOT NULL PRIMARY KEY,
    vaccine_schedule_description TEXT NOT NULL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_schedulerounds (
    scheduleround_id INTEGER NOT NULL PRIMARY KEY
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_target_population (
    target_pop_code TEXT NOT NULL PRIMARY KEY,
    target_pop_description TEXT NOT NULL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_geo_area (
    geo_area_code TEXT NOT NULL PRIMARY KEY
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_age_administered (
    age_administered_code TEXT NOT NULL PRIMARY KEY
);
''')

print("Dimension tables created successfully.")
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS fact_coverage (
    country_code TEXT NOT NULL,
    year_id INTEGER NOT NULL,
    antigen_code TEXT NOT NULL,
    coverage_category_code TEXT NOT NULL,
    target_number INTEGER NOT NULL,
    doses INTEGER NOT NULL,
    coverage REAL NOT NULL,
    PRIMARY KEY (country_code, year_id, antigen_code, coverage_category_code),
    FOREIGN KEY (country_code) REFERENCES dim_country(country_code),
    FOREIGN KEY (year_id) REFERENCES dim_year(year_id),
    FOREIGN KEY (antigen_code) REFERENCES dim_antigen(antigen_code),
    FOREIGN KEY (coverage_category_code) REFERENCES dim_coverage_category(coverage_category_code)
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS fact_incidence (
    country_code TEXT NOT NULL,
    year_id INTEGER NOT NULL,
    disease_code TEXT NOT NULL,
    denominator_description TEXT NOT NULL,
    incidence_rate REAL NOT NULL,
    PRIMARY KEY (country_code, year_id, disease_code, denominator_description),
    FOREIGN KEY (country_code) REFERENCES dim_country(country_code),
    FOREIGN KEY (year_id) REFERENCES dim_year(year_id),
    FOREIGN KEY (disease_code) REFERENCES dim_disease(disease_code),
    FOREIGN KEY (denominator_description) REFERENCES dim_denominator(denominator_description)
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS fact_reported_cases (
    country_code TEXT NOT NULL,
    year_id INTEGER NOT NULL,
    disease_code TEXT NOT NULL,
    cases REAL NOT NULL,
    PRIMARY KEY (country_code, year_id, disease_code),
    FOREIGN KEY (country_code) REFERENCES dim_country(country_code),
    FOREIGN KEY (year_id) REFERENCES dim_year(year_id),
    FOREIGN KEY (disease_code) REFERENCES dim_disease(disease_code)
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS fact_vaccine_introduction (
    country_code TEXT NOT NULL,
    year_id INTEGER NOT NULL,
    vaccine_intro_product_name TEXT NOT NULL,
    intro_status_code TEXT NOT NULL,
    PRIMARY KEY (country_code, year_id, vaccine_intro_product_name),
    FOREIGN KEY (country_code) REFERENCES dim_country(country_code),
    FOREIGN KEY (year_id) REFERENCES dim_year(year_id),
    FOREIGN KEY (vaccine_intro_product_name) REFERENCES dim_vaccine_intro_product(vaccine_intro_product_name),
    FOREIGN KEY (intro_status_code) REFERENCES dim_intro_status(intro_status_code)
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS fact_vaccine_schedule (
    country_code TEXT NOT NULL,
    year_id INTEGER NOT NULL,
    vaccine_schedule_code TEXT NOT NULL,
    scheduleround_id INTEGER NOT NULL,
    target_pop_code TEXT NOT NULL,
    geo_area_code TEXT NOT NULL,
    age_administered_code TEXT NOT NULL,
    source_comment TEXT,
    PRIMARY KEY (country_code, year_id, vaccine_schedule_code, scheduleround_id, target_pop_code, geo_area_code, age_administered_code),
    FOREIGN KEY (country_code) REFERENCES dim_country(country_code),
    FOREIGN KEY (year_id) REFERENCES dim_year(year_id),
    FOREIGN KEY (vaccine_schedule_code) REFERENCES dim_vaccine_schedule(vaccine_schedule_code),
    FOREIGN KEY (scheduleround_id) REFERENCES dim_schedulerounds(scheduleround_id),
    FOREIGN KEY (target_pop_code) REFERENCES dim_target_population(target_pop_code),
    FOREIGN KEY (geo_area_code) REFERENCES dim_geo_area(geo_area_code),
    FOREIGN KEY (age_administered_code) REFERENCES dim_age_administered(age_administered_code)
);
''')

conn.commit()

print("Fact tables created and changes committed successfully.")
```
